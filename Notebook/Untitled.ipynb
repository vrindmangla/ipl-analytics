```python
import pandas as pd

# Load data
df = pd.read_csv( r"C:\Users\LENOVO\Desktop\IPL Data Analytics Dashboard\Dataset\ipl_2026_deliveries.csv")

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Remove leading/trailing spaces from column names
df.columns = df.columns.str.strip()

# Convert date column
df['date'] = pd.to_datetime(df['date'])

# Fill missing values in numeric columns
numeric_cols = [
    'runs_of_bat', 'extras', 'wide',
    'legbyes', 'byes', 'noballs'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df[col].fillna(0)

# Fill missing values in categorical columns
categorical_cols = [
    'wicket_type',
    'player_dismissed',
    'fielder'
]

for col in categorical_cols:
    df[col] = df[col].fillna('Not Applicable')

# Standardize text columns
text_cols = [
    'venue', 'batting_team', 'bowling_team',
    'striker', 'bowler'
]

for col in text_cols:
    df[col] = df[col].str.strip()

# Create total runs scored on the ball
df['total_runs'] = df['runs_of_bat'] + df['extras']

# Create wicket flag
df['is_wicket'] = df['player_dismissed'].apply(
    lambda x: 0 if x == 'Not Applicable' else 1
)

# Check missing values
print(df.isnull().sum())

# Dataset information
print(df.info())

df.to_excel(
    r"C:\Users\LENOVO\Desktop\IPL Data Analytics Dashboard\Dataset\ipl_2026_deliveries_modified.xlsx",
    index=False
)

print("Excel file saved successfully!")
```
